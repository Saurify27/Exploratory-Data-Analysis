```python
datasheet_copy=datasheet.copy()
```

```python
datasheet_copy['transaction_dates']= pd.to_datetime(datasheet['extraction'])
```

```python
datasheet_copy['transaction_month']=datasheet_copy['transaction_dates'].dt.month
```

```python
datasheet_copy['transaction_day_name'] =datasheet_copy['transaction_dates'].dt.weekday_name
```

```python
No_of_transactions_by_every_customer = datasheet_copy.groupby(['customer_id']).count()
```

```python
print(No_of_transactions_by_every_customer['transaction_id'])
```

```python
mean_amount_of_transaction_for_each_month=datasheet_copy.groupby(['transaction_month']).mean()
```

```python
print(mean_amount_of_transaction_for_each_month['amount'])
```

```python
mean_amount_of_transaction_by_each_customer_for_each_month=datasheet_copy.groupby(['customer_id','transaction_month']).mean()
```

```python
print(mean_amount_of_transaction_by_each_customer_for_each_month['amount'])
```

```python
No_of_transactions_by_every_customer_in_each_month= datasheet_copy.groupby(['customer_id','transaction_month']).count()
```

```python
print(No_of_transactions_by_every_customer_in_each_month['transaction_id'])
```

```python
mean_transaction_amt_per_weekday= datasheet_copy.groupby(['transaction_day_name']).mean()
```

```python
print(mean_transaction_amt_per_weekday['amount'])
```

```python
plt.plot([8,9,10],mean_amount_of_transaction_for_each_month.amount)
plt.xlabel("Month")
plt.ylabel("Amount of Transaction (AUD)")
plt.show()
```

```python
salary_transactions= datasheet_copy[datasheet_copy.txn_description =='PAY/SALARY']
```

```python
salary_transactions
```

```python
salary_transactions.drop_duplicates(subset=['amount'])
```

```python
Annual_salary_list =pd.DataFrame(columns=['customer_id','age','weekly_salary','annual_salary','mean_purchase_amt_for_three_months','mean_yearly_purchase_amt'])
```

```python
Annual_salary_list.customer_id= salary_transactions.drop_duplicates(subset=['amount']).customer_id
```

```python
Annual_salary_list.age= salary_transactions.drop_duplicates(subset=['amount']).age
```

```python
Annual_salary_list.weekly_salary= salary_transactions.drop_duplicates(subset=['amount']).amount
```

```python
Annual_salary_list.annual_salary= Annual_salary_list.weekly_salary*52
```

```python
Annual_salary_list.set_index('customer_id',inplace=True)
```

```python
purchase_transactions= datasheet_copy[(datasheet_copy.txn_description== 'POS') | (datasheet_copy.txn_description== 'SALES-POS')]
purchase_transactions
```

```python
x=purchase_transactions.groupby('customer_id').amount.mean()
x
```

```python
Annual_salary_list.mean_purchase_amt_for_three_months=x
Annual_salary_list.mean_yearly_purchase_amt=x*4
Annual_salary_list
```

```python
plt.scatter(Annual_salary_list.mean_purchase_amt_for_three_months,Annual_salary_list.annual_salary,c='b')
plt.show()
```

```python
plt.scatter(Annual_salary_list.age,Annual_salary_list.annual_salary,c='r')
plt.show()
```
