The given Dataset is loaded into a dataframe using pandas library.


```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
excel_worksheet='ANZ synthesised transaction dataset.xlsx'
```

```python
sheet1=pd.read_excel('ANZ synthesised transaction dataset.xlsx')
```

```python
sheet1.head()
```

```python
sheet1.describe()
```

The Dataset contains columns with missing values.The columns with missing values are dropped and new dataset is loaded referred as 'datasheet'.

```python
datasheet=sheet1.dropna(axis=1)
datasheet
```

```python
datasheet.describe()
```

A copy of 'datasheet' is made to perform various operations upon. Furthermore, the transaction column is filtered and salary transactions are loaded into a new dataframe called 'salary_transactions'.

```python
datasheet_copy=datasheet.copy()
```

```python
salary_transactions= datasheet_copy[datasheet_copy.txn_description =='PAY/SALARY']
```

```python
salary_transactions
```

The redundant transactions are removed so that our dataframe contains only one salary transaction for each customer.

```python
salary_transactions.drop_duplicates(subset=['amount'])
```

A new dataframe called 'Annual_salary_list' is made in which several important details & insights are loaded as columns.

```python
Annual_salary_list =pd.DataFrame(columns=['customer_id','age','weekly_salary','annual_salary','mean_purchase_amt_for_three_months','mean_yearly_purchase_amt'])
```

```python
Annual_salary_list.customer_id= salary_transactions.drop_duplicates(subset=['amount']).customer_id
```

```python
Annual_salary_list.age= salary_transactions.drop_duplicates(subset=['amount']).age
```

```python
Annual_salary_list.weekly_salary= salary_transactions.drop_duplicates(subset=['amount']).amount
```

```python
Annual_salary_list.annual_salary= Annual_salary_list.weekly_salary*52
```

```python
Annual_salary_list.set_index('customer_id',inplace=True)
```

The original dataframe is filtered to find purchase transactions and are loaded in a new dataframe called 'purchase_transactions'.
Further the mean purchase amount for given 3 months by every customer is calculated.

```python
purchase_transactions= datasheet_copy[(datasheet_copy.txn_description== 'POS') | (datasheet_copy.txn_description== 'SALES-POS')]
purchase_transactions
```

```python
x=purchase_transactions.groupby('customer_id').amount.mean()
x
```

Here, the goal is to find a customer attribute based on their spending habits. So, the mean spending for 3 months is found and it is multiplied by 4 to make a new attribute called 'mean_yearly_purchase_amt' and it is added as a column to the 'Annual_salary_list' dataframe.

```python
Annual_salary_list.mean_purchase_amt_for_three_months=x
Annual_salary_list.mean_yearly_purchase_amt=x*4
Annual_salary_list
```

The Correlation between customer attribute 'age' and 'annual_salary' is plotted using a scatter plot. 

```python
plt.scatter(Annual_salary_list.age,Annual_salary_list.annual_salary,c='r')
plt.show()
```

Furthermore, a regression model is built to predict annual salaries based on previous observations of age and salaries.

```python
import numpy as np
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
```

```python
reg= linear_model.LinearRegression()
```

The 'age' column and 'annual_salary' column is split into train and test sets and fitted using the regression function.

```python
x_train,x_test,y_train,y_test = train_test_split(Annual_salary_list[['age']],Annual_salary_list.annual_salary,test_size=0.2,random_state=4)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

The salaries of test set observations are predicted using regression model and can be compared with actual salaries which are given below.

```python
predicted_salaries=reg.predict(x_test)
predicted_salaries
```

```python
y_test
```

Now that the regression model is built, its accuracy is checked using many factors,one of which is the mean squared error method in which the mean of squares of error(predicted value-actual value) is calculated. If the value is found to be high, the model is said to be poorly performing.

In case of our regression, the mean squared error value is very which suggest that our model is not accurately predicting values or the data fed to model shows no correlation. Hence, it is clear that this model cannot be used to predict real world values. 

```python
# mean squared error to test accuracy of regression model

np.mean((predicted_salaries-y_test)**2)
```

Furthermore, another regression model is built to predict salary based on derived customer attribute which is the mean spending over an year(mean spending over 3 months * 4)

A scatter plot for the same values is plotted below.

```python
plt.scatter(Annual_salary_list.mean_yearly_purchase_amt,Annual_salary_list.annual_salary,c='b')
plt.show()
```

The 'mean_yearly_purchase_amt' column and 'annual_salary' column is split into train and test sets and fitted using the regression function.

```python
x_train,x_test,y_train,y_test = train_test_split(Annual_salary_list[['mean_yearly_purchase_amt']],Annual_salary_list.annual_salary,test_size=0.2,random_state=4)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

The salaries of test set observations are predicted using regression model and can be compared with actual salaries which are given below.

```python
predicted_salaries=reg.predict(x_test)
predicted_salaries
```

```python
y_test
```

In case of this regression model too, the value of mean squared error is found to be very high and hence it is concluded that this model is poorly performing and the data on which model is built shows no or very less correlation.

Hence, both the models cannot be used for reporting purposes.
